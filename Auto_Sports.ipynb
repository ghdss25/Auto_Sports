## Análise dos dados da Planilha Auto Sports 

*** 1 - Total de Vendas por Ano ***

```python
total_vendas_ano = auto_sports.groupby('Ano')['ValorVenda'].sum().rename("Total de Vendas")

analise_vendas = pd.DataFrame({"Total de Vendas" : total_vendas_ano})  

analise_vendas["Total de Vendas"] = analise_vendas["Total de Vendas"].map(lambda x: "R$ {:,.2f}".format(x))

analise_vendas
```

*** 2 - Custo de Entrega do Veículo Por Fabricante ***

```python
## Resumo estatístico do Custo de Entrega do Veiculo por Fabricante 
print(auto_sports.describe())
```

```python
## Cálculo da média do custo de entrega por fabricante 
media_custo_fabricante = auto_sports.groupby('Fabricante')['CustoEntrega'].mean().rename("Média")

analise_custo_fabricante = pd.DataFrame({"Média" : media_custo_fabricante}) 

analise_custo_fabricante["Média"] = analise_custo_fabricante["Média"].map(lambda x: "R$ {:,.2f}".format(x))

analise_custo_fabricante
```

```python
## Valor Máximo e mínimo do Custo de Entrega  

maximo_custo = auto_sports['CustoEntrega'].max()
minimo_custo = auto_sports['CustoEntrega'].min() 

analise = pd.DataFrame({"Valor Máximo": [maximo_custo], "Valor Mínimo": [minimo_custo]})

analise                      
```

*** 3 - Custo de Mão de Obra Por Estado ***

```python
## Cálculo do Custo Mão de Obra por estado 
media_custo_mao_obra = auto_sports.groupby('Estado')['CustoMaoDeObra'].mean().rename("Média")

analise = pd.DataFrame({"Média" : media_custo_mao_obra}) 

analise["Média"] = analise["Média"].map(lambda x: "R$ {:,.2f}".format(x))

analise
```

* Total de Vendas por Estado 

```python
total_vendas_estado = auto_sports.groupby('Estado')['ValorVenda'].sum().rename("Total por Estado") 

analise = pd.DataFrame({"Total por Estado":total_vendas_estado}) 

analise["Total por Estado"] = analise["Total por Estado"].map(lambda x: "R$ {:,.2f}".format(x)) 

analise
```

* Matriz de Vendas 

```python
matriz_vendas = auto_sports.pivot_table(values='ValorVenda', index=['Fabricante', 'Estado', 'NomeCliente', 'Modelo', 'Cor'], aggfunc='sum').reset_index()

matriz_vendas.rename(columns={'ValorVenda': 'Total de Vendas'}, inplace=True)

matriz_vendas['Total de Vendas'] = matriz_vendas['Total de Vendas'].map(lambda x: "R$ {:,.2f}".format(x))

matriz_vendas = matriz_vendas.sort_values(by='Total de Vendas', ascending=False)

matriz_vendas
```

* Vendas Acima ou Abaixo da Média 

```python
media_vendas = auto_sports['ValorVenda'].mean()

def status_situacao(row): 
    
    if row['ValorVenda'] > media_vendas: 
        
        return 'Acima da Média' 
    
    elif row['ValorVenda'] < media_vendas: 
        
        return 'Abaixo da Média' 
    
    else: 
        
        return 'Igual á Média' 
    
auto_sports['Status_Situacao'] = auto_sports.apply(status_situacao, axis=1) 
display(auto_sports[['Estado', 'ValorVenda', 'Status_Situacao']].head(90))
```

```python
media_de_vendas = auto_sports.groupby(["Estado", "Status_Situacao"])["ValorVenda"].mean().rename("Situação") 

analise_situacao = pd.DataFrame({"Situação" : media_de_vendas}) 

analise_situacao["Situação"] = analise_situacao["Situação"].map(lambda x: "R$ {:,.2f}".format(x)) 

analise_situacao
```

* Evolução das Vendas da Marca Jaguar por Ano e por Estado

```python
vendas_jaguar = auto_sports[auto_sports['Fabricante'] == 'Jaguar'] 

vendas_jaguar_por_estado_ano = vendas_jaguar.groupby(['Estado', 'Ano'])['ValorVenda'].sum().reset_index() 

vendas_jaguar_por_estado_ano['ValorVenda'] = vendas_jaguar_por_estado_ano['ValorVenda'].map(lambda x: "R$ {:,.2f}".format(x))

vendas_jaguar_por_estado_ano
```
