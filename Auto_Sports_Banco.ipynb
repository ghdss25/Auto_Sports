## Consulta Total de Vendas por Estado 

```python
consulta_total_vendas = pd.read_sql_query('SELECT Estado, AVG(ValorVenda) AS TotalVendas FROM fato_auto_sports GROUP BY Estado', 
                                          con=dbconn)
consulta_total_vendas['TotalVendas'] = consulta_total_vendas['TotalVendas'].apply(lambda x: locale.currency(x, grouping=True))

consulta_total_vendas
```

## Consulta da Média nas Vendas

```python
consulta_media_vendas = pd.read_sql_query('SELECT AVG(ValorVenda) AS MediaVendas FROM fato_auto_sports', 
                                          con=dbconn)

consulta_media_vendas['MediaVendas'] = consulta_media_vendas['MediaVendas'].apply(lambda x: locale.currency(x, grouping=True))

consulta_media_vendas
```

## Cálculo Acima da Média ou Abaíxo da Média 

```python
# Cálculo Acima da Média ou Abaixo da Média

query = '''
    SELECT fa.Estado, 
           AVG(fa.ValorVenda) AS Media_Vendas,
           CASE WHEN AVG(fa.ValorVenda) > media.Media_Geral THEN 'Acima da Média'
                ELSE 'Abaixo da Média'
           END AS StatusVenda
    FROM fato_auto_sports AS fa
    CROSS JOIN (
        SELECT AVG(ValorVenda) AS Media_Geral
        FROM fato_auto_sports
    ) AS media
    GROUP BY fa.Estado, media.Media_Geral
'''

situacao_acima_abaixo = pd.read_sql_query(query, con=dbconn)

situacao_acima_abaixo['Media_Vendas'] = situacao_acima_abaixo['Media_Vendas'].apply(lambda x: locale.currency(x, grouping=True))

situacao_acima_abaixo
```
